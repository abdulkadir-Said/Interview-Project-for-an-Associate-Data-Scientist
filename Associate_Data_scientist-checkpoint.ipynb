###### KNN

The Knearest Neighbor is robust to noisy data and it's an effective algorithm in cases where we have a large number of training data


######  For this algorithm we need to determine:
* Parameter K (number of nearest neighbors)
* type of distance (Euclidean distance, cosine similarity measure, Minkowsky, correlation, and Chi square)
* Computation time (takes time as we need to compute distance of each query instance to all training samples)

###### Random Forest

This is an ensemble learner (also known as weak learner). It utilizes a number of decison trees. You set the number of decison tree you want. Then a vote will happen to decide what category an observation is. Then the most voted for category will be used as an ouput.
